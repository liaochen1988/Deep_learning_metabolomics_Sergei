```python
import cobra
import pandas as pd
```

```python
df_bigg_rxns = pd.read_csv("bigg_models_reactions.txt", sep="\t")
reactions2remove = []
for rxn_id, rxn_str in zip(df_bigg_rxns.bigg_id, df_bigg_rxns.reaction_string):
    remove = False
    for _str in rxn_str.split(' '):
        _str2 = _str.strip()
        try:
            float(_str2) # stoichiometric coefficient
        except:
            if _str2 in ["<->","","+"]:
                continue
            else:
                assert "_" in _str2
                if _str2.split("_")[-1] in ["c","p","e"]:
                    continue
                else:
                    remove=True
                    break
    if remove:
        reactions2remove.append(rxn_id)
print("number of reactions removed = %d."%(len(reactions2remove)))
```

```python
bigg = cobra.io.read_sbml_model("bigg_universe.xml")
```

```python
print("number of reactions before removing: %d"%(len(bigg.reactions)))
bigg.remove_reactions(reactions2remove, remove_orphans=True)
print("number of reactions after removing: %d"%(len(bigg.reactions)))
```

```python
cobra.io.write_sbml_model(bigg, "bigg_universe_compartment_cpe.xml")
```
