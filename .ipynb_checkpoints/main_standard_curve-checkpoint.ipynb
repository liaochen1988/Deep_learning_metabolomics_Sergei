```python
import pandas as pd
import numpy as np
from collections import Counter
from scipy.stats import ttest_1samp
from statsmodels.stats.multitest import multipletests
import os
import matplotlib.pyplot as plt
import seaborn as sns
```

# Read metabolomics data

```python
df_fc = pd.read_excel('../Original_Data/in_vitro_data.xlsx', sheet_name='foldchange.dmrvf.fa.ps', index_col=0)
df_fc.index.name = 'EntryID'
df_fc = df_fc.stack().reset_index()
df_fc.columns = ['EntryID', 'dname', 'FC']
assert df_fc.FC.min()>0
df_fc.head()
```

# Read experimental information

```python
df_info = pd.read_excel('../Original_Data/in_vitro_data.xlsx', sheet_name='aggregated_md').rename({'Unnamed: 0':'EntryID'}, axis=1)
df_info.head()
```

# Join the two tables

```python
df = pd.merge(df_fc, df_info[['EntryID','sample_type','media','taxonomy']], left_on=['EntryID'], right_on=['EntryID'], how='left')
df = df[df.taxonomy.notnull()]
assert len(set(df.sample_type)) == 1 and list(set(df.sample_type))[0]=='supernatant'
df = df.drop(['EntryID','sample_type'], axis=1)
df.head()
```

# Convert to bigg id

```python
pd.DataFrame.from_dict(dict(Counter(df.media)), orient='index').sort_values(0, ascending=False)
```

```python
df_bigg = pd.read_csv('../Convert_Metabolite_IDs/Metabolite_DB_ids_w_biggids.csv')
df_bigg = df_bigg[df_bigg.universal_bigg_id.notnull()]
df2 = pd.merge(df, df_bigg[['dname','universal_bigg_id']], left_on=['dname'], right_on=['dname'], how='inner')
df2 = df2[df2.media=='mm'] # we will focus on mm media
df2['log2FC'] = np.log2(df2['FC'])
df2 = df2.drop(['dname','media','FC'], axis=1)
df2.head()
```

# Make sure that taxonomy in df2 can be found in metadata

```python
df_meta = pd.read_csv('../strain_model_metadata.csv')
assert len(set(df2.taxonomy) - set(df_meta.taxonomy))==0
df3 = pd.merge(df2, df_meta[['genome_id','taxonomy']], left_on=['taxonomy'], right_on=['taxonomy'], how='left').drop('taxonomy', axis=1)
df3.head()
```

# Compute statistics of metabolite fold change

```python
lines = []
all_strains = set(df3.genome_id)
all_metabolites = set(df3.universal_bigg_id)
for strain in all_strains:
    for met in all_metabolites:
        df_tmp = df3[(df3.genome_id==strain) & (df3.universal_bigg_id==met)]
        if len(df_tmp)>0:
            n_reps = len(df_tmp)
            mean_log2fc = np.mean(df_tmp.log2FC)
            var_log2fc = np.var(df_tmp.log2FC)
            p = ttest_1samp(list(df_tmp.log2FC), 0.0)[1]
            lines.append([strain, met, n_reps, mean_log2fc, var_log2fc, p])
df4 = pd.DataFrame(lines, columns=['genome_id','universal_bigg_id','n_reps','log2fc_mean','log2fc_var','log2fc_P'])
df4 = df4[df4.log2fc_P.notnull()]
df4['log2fc_Padj'] = multipletests(df4.log2fc_P, method='fdr_bh')[1]
df4.to_csv('metabolites_log2fc_stats.csv', index=False)
df4.head()
```

# Filter for significant metabolties (sig if adj. P<0.05)

```python
df5 = pd.read_csv('metabolites_log2fc_stats.csv')
df5 = df5[df5.log2fc_Padj<0.05]
df5
```

# Metabolites that are consumed must be present in the mega medium

```python
# read metabolite information
df_bigg2 = pd.read_csv('../BiggDatabase/bigg_models_metabolites.txt', sep='\t')
df_bigg2.head()
```

```python
consumed_metabolites = list(set(df5[df5.log2fc_mean<0].universal_bigg_id))
other_supplements = ['cl','fe3','fe2','mn2','mg2','ca2','k','pi','na1','cu2','modb','ni2','zn2','h2co3','h','h2o','so4','cobalt2','no3','sel', # trace minerals
                     'ac','but','ppa','ival', # scfa supplements
                     'cys__L', # L-Cysteine hydrochloride
                     'glc__D','cellb','malt','fru', # carbon sources
                     'his__L', # Histidine-Hematin
                     'mndn', # vitamin K
                     'fol','pydxn','ribflv','btn','thm','nac','pnto__R','cbl1','4abz','lipt' # ATCC Vitamin supplements
                    ]
df_mm = pd.DataFrame(list(set(consumed_metabolites+other_supplements))).rename({0:'compound'}, axis=1)
df_mm['medium'] = 'mm'
df_mm = pd.merge(df_mm, df_bigg2[['universal_bigg_id','name']].drop_duplicates().rename({'universal_bigg_id':'compound'}, axis=1), left_on=['compound'], right_on=['compound'], how='left')
df_mm['description'] = 'mega medium, preparation protocol'
df_mm = df_mm[['medium','description','compound','name']]
df_mm.loc[df_mm.compound.isin(consumed_metabolites),'description'] = 'mega medium, metabolomics'
df_mm.to_csv('../Culture_media/mm.tsv', sep='\t', index=False)
df_mm
```

# Write soft constraints

```python
df_bigg_rxn = pd.read_csv('../BiggDatabase/bigg_models_reactions.txt', sep='\t')
all_bigg_rxns = set(df_bigg_rxn.bigg_id)
```

```python
all_genomes = set(df5.genome_id)
for genome in all_genomes:
    df5_tmp = df5[df5.genome_id == genome]
    consumed_metabolites = set(df5_tmp[df5_tmp.log2fc_mean<0].universal_bigg_id)
    secreted_metabolites = set(df5_tmp[df5_tmp.log2fc_mean>0].universal_bigg_id)
    consumption_rxns = ['EX_%s_e'%(met) for met in consumed_metabolites if 'EX_%s_e'%(met) in all_bigg_rxns]
    secretion_rxns = ['EX_%s_e'%(met) for met in secreted_metabolites if 'EX_%s_e'%(met) in all_bigg_rxns]
    lines = []
    for rxn in consumption_rxns:
        lines.append(['R_'+rxn, -1]) # reaction identifier, reaction direction (1 for forward, -1 for backward, 0 does not occur)
    for rxn in secretion_rxns:
        lines.append(['R_'+rxn, +1])
    df2write = pd.DataFrame(lines, columns=['reaction','value'])
    df2write.to_csv('%s_soft_constraints.tsv'%(genome), sep='\t', index=False)
```

# Run carveme with iterative hard constraints (iterative)

```python
for f in os.listdir('../Genomes/'):
    if f.endswith('.fna'):
        genome=f.replace('.fna', '')
        print(genome)
        
        if os.path.isfile("../GSM_reconstruction/hard_constraints_iterative/%s_hard.xml"%(genome)):
            continue
        
        # select metabolites that show significant change and rank them by their fold change (from low to high in absolute value)
        df5_tmp = df5[df5.genome_id == genome]
        df5_tmp = df5_tmp.reindex(df5_tmp.log2fc_mean.abs().sort_values().index).reset_index(drop=True)

        # find the maximum consistency
        for start_index in np.arange(0,len(df5_tmp)):
    
            # write hard constraints
            lines = []
            for met, log2fc in zip(df5_tmp.iloc[start_index:].universal_bigg_id, df5_tmp.iloc[start_index:].log2fc_mean):
                if log2fc<0:
                    lines.append(['R_EX_%s_e'%(met), -1000, -0.001]) # reaction identifier, lower bound, upper bound
                else:
                    lines.append(['R_EX_%s_e'%(met), +0.001, +1000]) # reaction identifier, lower bound, upper bound
            
            # anaerobic condition
            lines.append(['R_EX_o2_e',0,0])
            
            df2write = pd.DataFrame(lines, columns=['reaction','lower_bound', 'upper_bound'])
            df2write.to_csv('%s_hard_constraints.tsv'%(genome), sep='\t', index=False)
                
            # run carveme
            cmd = "carve \"../Prokka_annotation/%s/%s.faa\" --hard \"%s\" --init \"mm\" --gapfill \"mm\" --mediadb \"../Culture_media/mm.tsv\" --output \"../GSM_reconstruction/hard_constraints_iterative/%s_hard.xml\""%(genome,genome,'%s_hard_constraints.tsv'%(genome),genome)
            okay = os.system(cmd)
            
            if okay==0:
                break
```

# Update meta data

```python
lines = []
for f in os.listdir('../GSM_reconstruction/hard_constraints_iterative/'):
    if f.endswith("_hard.xml"):
        genome = f.rstrip('_hard.xml')
        n_constraints_total = len(df5[df5.genome_id == genome])
        df_hard_constraints = pd.read_csv('%s_hard_constraints.tsv'%(genome), sep='\t')
        df_hard_constraints = df_hard_constraints[df_hard_constraints.reaction != 'R_EX_o2_e']
        n_constraints_used = len(df_hard_constraints)
        lines.append([genome, n_constraints_total, n_constraints_used])
df_model_info = pd.DataFrame(lines, columns=['genome_id', 'cstr_num_total', 'cstr_num_used'])
print('number of models = %d'%(len(df_model_info)))
df_model_info.head()
```

```python
len(set(df_model_info.genome_id))
```

```python
# plot number of evidence vs implemented constraints
df2plot = df_model_info.sort_values('cstr_num_total', ascending=True).set_index('genome_id')
df2plot['cstr_num_not_used'] = df2plot['cstr_num_total']-df2plot['cstr_num_used']

fig, ax = plt.subplots(figsize=(25,5))
_ = df2plot[['cstr_num_used','cstr_num_not_used']].plot(kind='bar', stacked=True, ax=ax, width=0.75, cmap='Set2')

plt.savefig('hard_constraints.png')
```

```python
df_meta = pd.read_csv('../strain_model_metadata.csv')
df_meta = pd.merge(df_meta, df_model_info, left_on='genome_id', right_on='genome_id', how='left')
df_meta.to_csv('../strain_model_metadata.csv', index=False)
df_meta.head()
```
