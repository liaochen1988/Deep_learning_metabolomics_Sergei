# Self defined functions

```python
def Zwietering_Logistic_growth_model(x, y0, A, lag, mu):
    return  y0+(A-y0)/(1+np.exp(4*mu/A*(lag-x)+2))

def get_r2_coefficient(function, popt, x, y):
    ypred = function(x, *popt)
    _, _, r_value, _, _ = linregress(ypred, y)
    r_squared = r_value ** 2
    return r_squared

def extract_growth_rate(x,y):
    while 1:
        best_fits, covar = curve_fit(Zwietering_Logistic_growth_model,
                                     x,
                                     y,
                                     p0=[np.min(y), random.random()*(np.max(y)-np.min(y))+np.min(y), random.random()*np.max(x), random.random()],
                                     bounds = ([np.min(y), np.min(y), 0, 0], [np.max(y), np.max(y), np.max(x), np.inf]),
                                     maxfev=10000
                                    )
        r2 = get_r2_coefficient(Zwietering_Logistic_growth_model, best_fits, x, y)
        if r2>0.9:
            break

    return best_fits, r2
```

# Extract maximum growth rate

```python
growth_rate = []
for sid in df_od.sample_id:
    time_step = list(df_od.loc[df_od.sample_id==sid, 'od_time_step'])[0]
    df2 = df_od.loc[df_od.sample_id==sid, [c for c in df_od.columns if c.startswith('m') and c != 'media']].T
    df2.columns = ['OD']
    df2 = df2.dropna()
    df2 = df2.rename({idx:int(idx.lstrip('m'))*time_step for idx in df2.index})
    
    # manually select time window for parameter fitting
    if sid=='od00447':
        df2 = df2[df2.index<80]
    if sid in ['od00468','od00492','od00177']:
        df2 = df2[df2.index<40]
    if sid in ['od00350']:
        df2 = df2[df2.index<30]
    if sid in ['od00505','od00140','od00479','od00735','od00769','od00722']:
        df2 = df2[df2.index<20]
    if sid in ['od00602']:
        df2 = df2[df2.index<15]
    if sid in ['od00525','od00001','od00326']:
        growth_rate.append([sid, 0, np.nan])
        continue
    
    x = list(df2.index)
    y = list(df2.OD)
    yhat = list(savgol_filter(y, 51, 3)) # smoothing
    if np.max(yhat)/np.min(yhat)<1.10:
        # do not grow
        growth_rate.append([sid, 0, np.nan])
    else:
        # fit growth curve only up to the peak
        max_ind = yhat.index(np.max(yhat))
        xtrunc = x[0:max_ind+1]
        ytrunc = yhat[0:max_ind+1]
        if len(xtrunc)==1: # od curve is decreasing
            growth_rate.append([sid, 0, np.nan])
        else:
            bfit, r2 = extract_growth_rate(xtrunc, ytrunc)
            ypred = Zwietering_Logistic_growth_model(x, *bfit)
            growth_rate.append([sid, bfit[3], r2])
            assert r2>0.9

df_growth = pd.DataFrame(growth_rate, columns=['sample_id', 'max_growth', 'r2'])
df_growth = pd.merge(df_od[["sample_id","taxonomy","experiment","od_loc_well","preculture_date","preculture_time","subculture_time","media"]],
                     df_growth, left_on='sample_id', right_on='sample_id', how='inner')
df_growth.to_csv('Data/han_nature_computed_max_growth_rate.csv', index=False)
df_growth.head()
```
